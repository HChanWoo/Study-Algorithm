### 조건문

### 조건문

```python
H, M = map(int, input().split())
timer = int(input()) 

H += timer // 60
M += timer % 60

if M >= 60:
    H += 1
    M -= 60
if H >= 24:
    H -= 24

print(H,M)
```

### 2480

```python
list_dice = [0,0,0,0,0,0]
list_roll = list(map(int,input().split()))
for i in range(3) :
    list_dice[list_roll[i]-1]+=1

if 3 in list_dice :
    print(10000+(list_dice.index(3)+1)*1000)
elif 2 in list_dice :
    print(1000+(list_dice.index(2)+1)*100)
else :
    print((6-list_dice[::-1].index(1))*100)
```

### 8958

```python
import sys
input = sys.stdin.readline

N = int(input())
for i in range(N) :
    total = 0
    score = 1
    test_case = input().strip()
    for st in test_case :
        if st == 'X' :
            score=1
        else :
            total+=score
            score+=1
    print(total)
```

### 4344

```python
N = int(input())
for i in range(N) :
    score = list(map(int,input().split()))
    avg = sum(score[1:])/score[0]
    
    up = 0
    for j in score[1:] :
        if j > avg :
            up+=1
    rate = up/score[0]*100
    print(f'{rate:.3f}%')
```

### 1157

```python
# A : 65 / a : 97 -65 /32
N = input()
list_num = [0]*26

for i in N :
    list_num[(ord(i)-65)%32]+=1

max_num = max(list_num)
max_idx = list_num.index(max(list_num))

if max_num in list_num[max_idx+1:]  :
    print('?')
else :
    print(chr(max_idx+65))
```
