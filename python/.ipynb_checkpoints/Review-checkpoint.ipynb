### 4344

```python
N = int(input())
for i in range(N) :
    score = list(map(int,input().split()))
    avg = sum(score[1:])/score[0]
    
    up = 0
    for j in score[1:] :
        if j > avg :
            up+=1
    rate = up/score[0]*100
    print(f'{rate:.3f}%')
```

### 2292

```python
find = int(input())
N=2
list_room = [i for i in range(2,8)]
last_num = list_room[-1]

if find not in list_room :
    while(True) :
        N+=1
        list_room = [i for i in range(last_num+1,last_num+6*(N-1)+1)]

        if find in list_room :
            break
        last_num = list_room[-1]
    
print(N)
```

### 1193

```python
cnt = 1
cnt_sum = 1

find_num = int(input())

while(True) :
    if cnt_sum >= find_num :
        break
    cnt+=1
    cnt_sum+=cnt

if cnt==1 :
    idx=1
else :
    idx = find_num - cnt_sum + cnt
if cnt%2==1 :
    print('%d/%d'%((cnt-idx+1),idx))
else :
    print('%d/%d'%(idx,(cnt-idx+1)))
```

### 2839

```python
N = int(input())

five = N//5
three = N%5/3

if three != int(three) :
    
    for i in range(1,6) :
        if N-3*i < 0 :
            i=5
            break
        if (N-3*i)%5 == 0 :
            break
    if i==5 :
        print(-1)
    else :
        print((N-3*i)//5+i)
else :
    print(int(five+three))
```

## 기본수학2

### 2775

```python
M = int(input())
N = int(input())

list_so = list()
for num in range(M,N+1) :
    check = 0
    if num>1 :
        for i in range(2,num) :
            if num%i == 0 :
                check=1
                break
        if check == 0:
            list_so.append(num)

if len(list_so) == 0 :
    print(-1)
else :
    print('%d\n%d'%(sum(list_so),min(list_so)))
```

======= 이상적인 답 ========

```python
M = int(input())
N = int(input())

# 0,1 을 제외
list_total = [False,False] + [True] * 9999

for i in range(2,101) :
    if list_total[i] :
        for j in range(2*i,len(list_total),i) :
            list_total[j]=False

list_so = [i for i in range(M,N+1) if list_total[i]]

if len(list_so) == 0 :
    print(-1)
else :
    print('%d\n%d'%(sum(list_so),min(list_so)))
```
