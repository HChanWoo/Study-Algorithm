# 큐,덱 1966

```python
from collections import deque

cnt_tmp = int(input())

for _ in range(cnt_tmp) :
    cnt_doc,target = map(int,input().split())
    list_score = deque(list(map(int,input().split())))
    
    list_position = deque([0 for _ in range(cnt_doc)])
    list_position[target] = 1
    cnt = 0
    
    while True :
        if list_score[0] == max(list_score) :
            cnt+=1
            if list_position[0] == 1:
                print(cnt)
                break
            else :
                list_score.popleft()
                list_position.popleft()
        else :
            list_score.append(list_score.popleft())
            list_position.append(list_position.popleft())
```

# 큐,덱 10866

```python
from collections import deque

cnt_cmd = int(input())
list_tmp = deque(list())

for i in range(cnt_cmd) :
    input_cmd = input().split()
    
    if input_cmd[0] == 'push_front' :
        list_tmp.appendleft(input_cmd[1])
        
    elif input_cmd[0] == 'push_back' :
        list_tmp.append(input_cmd[1])
        
    elif input_cmd[0] == 'pop_front' :
        if len(list_tmp) == 0 :
            print(-1)
        else :
            print(list_tmp.popleft())
        
    elif input_cmd[0] == 'pop_back' :
        if len(list_tmp) == 0 :
            print(-1)
        else :
            print(list_tmp.pop())
        
    elif input_cmd[0] == 'size' :
        print(len(list_tmp))
        
    elif input_cmd[0] == 'empty' :
        if len(list_tmp) == 0 :
            print(1)
        else :
            print(0)
            
    elif input_cmd[0] == 'front' :
        if len(list_tmp) == 0 :
            print(-1)
        else :
            print(list_tmp[0])
        
    else :
        if len(list_tmp) == 0 :
            print(-1)
        else :
            print(list_tmp[len(list_tmp)-1])
```

# 큐,덱 1021

```python
from collections import deque

size_list, cnt_tmp = map(int,input().split())
list_pick = deque(list(map(int,input().split())))

list_deque = deque([i for i in range(1,size_list+1)])
cnt=0

while list_pick :
    if list_deque[0] == list_pick[0] :
        list_deque.popleft()
        list_pick.popleft()
    
    else :
        if list_deque.index(list_pick[0])/len(list_deque) > 0.5 :
            list_deque.appendleft(list_deque.pop())
        else :
            list_deque.append(list_deque.popleft())
        cnt+=1
        
print(cnt)
```

```python
n, m = map(int, input().split())
s = list(map(int ,input().split()))
q = [i for i in range(1, n + 1)]
cnt = 0
for i in range(m):
    q_len = len(q)
    q_index = q.index(s[i])
    if q_index < q_len - q_index:
        while True:
            if q[0] == s[i]:
                del q[0]
                break
            else:
                q.append(q[0])
                del q[0]
                cnt += 1
    else:
        while True:
            if q[0] == s[i]:
                del q[0]
                break
            else:
                q.insert(0, q[-1])
                del q[-1]
                cnt += 1
print(cnt)
```

# 큐,덱 5430

```python
from collections import deque

cnt_test = int(input())

for i in range(cnt_test) :
    fun = input()
    len_list = int(input())
    list_tmp = input()
    if list_tmp != '[]' :
        list_tmp = deque(list(map(int,list_tmp.replace('[','').replace(']','').split(','))))
    error = 0
    
    for st in fun :
        if st == 'R' :
            list_tmp.reverse()
        else :
            if len(list_tmp) == 0 or list_tmp=='[]':
                error = 1
                break
            else :
                list_tmp.popleft()
        
    if error == 1 :
        print('error')
    else :
        print(list(list_tmp))
```

# 동적 계획법 1463

```python
n = int(input())
d = [0] * (n + 1)

for i in range(2, n + 1):

    d[i] = d[i - 1] + 1
    if i % 3 == 0:
        d[i] = min(d[i], d[i // 3] + 1)	
    if i % 2 == 0:
        d[i] = min(d[i], d[i // 2] + 1)
        
print(d[n])
```

# 이분 탐색 1920

```python
size_list = int(input())
list_search = sorted(list(map(int,input().split())))
size_target = int(input())
list_target = list(map(int,input().split()))

for target in list_target :
    
    start=0
    end = size_list-1
    while True :
        if start<=end :
            print(0)
            break
        
        mid = (start+end)//2
        
        if list_search[mid] == target :
            print(1)
            break
        elif list_search[mid] > target :
            end = mid-1
        else :
            start = mid+1
        
```
